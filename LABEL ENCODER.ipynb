```python
import pandas as pd
import numpy as np
import time
start_time = time.time()
```

```python
colors = ['blue','purple','purple','yellow','red','blue']
```

```python
def encode(list):
    code=0
    dict={}
    for i in range (len(list)):
        if dict.keys()==list[i]:
            continue
        else:
            dict[list[i]]=code
            code=code+1
    return dict
```

```python
show=encode(colors)
show
```
